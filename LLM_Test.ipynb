```python
import mlflow
import mlflow.sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load data
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2)

# Train a model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Log the run
with mlflow.start_run():
    mlflow.log_param("n_estimators", 100)
    mlflow.log_param("random_state", 42)
    accuracy = model.score(X_test, y_test)
    mlflow.log_metric("accuracy", accuracy)

    # Log the model
    mlflow.sklearn.log_model(model, "random_forest_model")
```

```python
import pandas as pd
```

```python
input = pd.read_pickle("machine_translation.pkl")
input_easy = input[input['complexity'] == 'easy']
input_easy
```

```python
output = llm_llama32("Translate this german text into english: \n\n" + input_easy['text_german'][0] + '\n\nTranslation:', max_tokens=None, echo=False)
translated_text = output['choices'][0]['text']
print(translated_text)
```

```python
input_easy['text_english_trans'] = ""
input_easy.loc[0, 'text_english_trans'] = translated_text.split('Note:')[0]
input_easy
```

```python
print(input_easy['text_english_trans'][0])
print(input_easy['text_english'][0])
```
